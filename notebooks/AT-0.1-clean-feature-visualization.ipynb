# Clean Feature Visualization

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
sns.set(context='notebook')
sns.set_palette("Set1")
sns.set_style("white")
```

```python
df = pd.read_csv("../data/clean/features_clean.csv")
df = df.drop('Unnamed: 0', axis=1)
df_X = df.drop(['Trump_tweet_id', 'user_id'], axis=1)
df_X.drop(['age', 'gen_prob_Other'], axis=1, inplace=True)
```

```python
n = df.shape[0]
print("Number of examples: {}".format(n))
```

```python
df_X.hist(figsize=(20,15), bins=25)
plt.show()
```

```python
corr = df_X.corr()
sns.heatmap(df_X.corr())
plt.show()
```
